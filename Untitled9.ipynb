```python
import numpy as np
import cv2
from sklearn.preprocessing import normalize

def predict_anime(uploaded_img_path, model, neighbours, y):
    img = cv2.imread(uploaded_img_path)
    img = cv2.resize(img, (224, 224))
    img = np.array(img)
    cv2_imshow(img)

    expand_img = np.expand_dims(img, axis=0)
    pre_img = preprocess_input(expand_img)
    result = model.predict(pre_img).flatten()
    normalized = result / np.linalg.norm(result)
    normalized = normalized.reshape(1, -1)
    distances, indices = neighbours.kneighbors(normalized, n_neighbors=6)
    guessed_anime = y[indices[0][0]]
    recommended_anime = [y[idx] for idx in indices[0] if y[idx] != guessed_anime][:5]

    return guessed_anime, recommended_anime
```
